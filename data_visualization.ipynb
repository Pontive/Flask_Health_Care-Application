```python
import pandas as pd
import matplotlib.pyplot as plt

# File path
file_path = r"C:\Users\YD\survey_results.csv"

# Load CSV into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
print(df.head())
```

```python
# Count of each gender
gender_counts = df['gender'].value_counts()

# Plot bar chart
plt.figure(figsize=(8, 6))
gender_counts.plot(kind='bar', color=['skyblue', 'orange'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Sum up expenses
expense_columns = ['utilities', 'entertainment', 'school_fees', 'shopping', 'healthcare']
expense_sums = df[expense_columns].sum()

# Plot pie chart
plt.figure(figsize=(8, 8))
expense_sums.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0'])
plt.title('Expense Distribution')
plt.ylabel('')  # Remove y-axis label
plt.show()
```

```python
# Plot boxplot for income
plt.figure(figsize=(8, 6))
df['total_income'] = pd.to_numeric(df['total_income'], errors='coerce')  # Ensure numeric data
df.boxplot(column='total_income', grid=False)
plt.title('Total Income Distribution')
plt.ylabel('Income')
plt.show()
```

```python
# Plot histogram for age
plt.figure(figsize=(8, 6))
df['age'] = pd.to_numeric(df['age'], errors='coerce')  # Ensure numeric data
df['age'].plot(kind='hist', bins=10, color='purple', alpha=0.7)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
